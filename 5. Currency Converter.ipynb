```python
# Put it in a while loop, so as to give the user the opportunity to try again(i.e, repetition) if he first of all gave an invalid input
while True:
    try:
        # Ask the user for an amount
        amount = float(input("Enter the amount: ")) # As we want to convert whatever the user enters here into a float number, as it is a number we're dealing with here.
        # First of all '0' or any other negative number shouldn't be considered as a valid input here at all.
        if amount <= 0:
            raise ValueError() # A better line of code than to just duplicate with "print('Invalid error')"
        # Incase the input was valid(that is a valid number been given as the input this time), break out the loop
        break
        # But if the about input given can't be converted into a 'float', we'll have a 'ValueError' here which will need to address differently, by eliminating it. This is why we'll need to put the try...except in the input.
    except ValueError:    
        print("Invalid amount")
# Now we conviniently have our valid amount        

currencies = ('CFA', 'USD', 'CAD')

# Loop, since we want to retry again
while True:        
    # Ask for the source currency
    source_currency = input("Source Currency(CFA/USD/CAD): ").upper() # As the '.upper()' accepts in lower case(e.g cfa) forms as well.
    # To validate this source currency, we can have a tuple do the checking for us. But this so called tuple, has to already be established 'outside our loop'. We can called it 'currencies'.
    # 'currecies' should be specifically outside the loop because, it is not a piece of logic that should 'not' be 'repeated'.
    # If invalid
    #    Print an error
    if source_currency not in currencies:
        print("Invalid Currency")
    else:
        break # Meaning a valid currency was been chosen, meaning we can break out of our loop now
    

# Loop as same thing goes, for the target currency. We'll simply repeat the same structure here.
while True:    
    # Ask for the target currency
    target_currency = input("Target Currency(CFA/USD/CAD): ").upper()    
    # If invalid
    #    Print an error
    if target_currency not in currencies:
        print("Invalid Currency")
    else:
        break 

# Do the conversion
# We'll use but a dictionary here to ease the work, as we'll be mapping every key to a value.
exchange_rates = {
    'CFA': {'USD': 0.0016, 'CAD': 0.0023},
    'USD': {'CFA': 623.3427, 'CAD': 1.416},
    'CAD': {'CFA': 439.647, 'USD': 0.70564}
}  

# But to avoid the 'KeyError' for the case of our source and target currencies being directly the same,
if source_currency == target_currency:
    converted_amount = amount
else:
    converted_amount = amount * exchange_rates[source_currency][target_currency]
    
# Finally, Print the result
print(f"{amount} {source_currency} is equal to {converted_amount:.2f}{target_currency}") # i.e, keeping our final result to just 2 decimal places.


```
